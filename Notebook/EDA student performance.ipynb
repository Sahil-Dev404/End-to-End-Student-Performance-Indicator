```python
#checking the all the unique categories mentioned in each column 
print("categories in the gender column",df["gender"].unique())
print("categories in the race ethnicity",df["race_ethnicity"].unique())
print("categories in the parental level of education",df["parental_level_of_education"].unique())
print("categories in the lunch",df["lunch"].unique())
print("categories in the test prep course",df["test_preparation_course"].unique())
```

```python
#defining the numeric and the cateogrical features in the data 
numeric_features=[feature for feature in df.columns if df[feature].dtype!="O"]
categorical_features=[feature for feature in df.columns if df[feature].dtype=="O"]
print("No. of numeric features:", len(numeric_features),"->",numeric_features)
print("No. of categorical features:", len(categorical_features),"->",categorical_features)
```

adding the columns for total score and average score

```python
df["total_score"]=df["math_score"]+df["reading_score"]+df["writing_score"]
df["avg_score"]=df["total_score"]/3
df.head(2)
```

```python
#checking the number of students that got the full marks in respective subjects
math_full=df[df["math_score"]==100]["avg_score"].count()
reading_full=df[df["reading_score"]==100]["avg_score"].count()
writing_full=df[df["writing_score"]==100]["avg_score"].count()

print("No. of students with full score in maths:",math_full)
print("No. of students with full score in reading:",reading_full)
print("No. of students with full score in writing:",writing_full)
```
